```python
"""
Exercise 2: Image Transposition
Using the image loaded in Exercise 1:
1. Transpose the image using NumPy.
2. Display the original and transposed images side by side.
3. Print the shapes of:
> Original image
> Transposed image
"""

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image



# Load the image (replace with your actual image path)
image = Image.open("/content/sample_data/Face.jpg")   # or whatever the path is
face_array = np.array(image)           # Convert PIL Image to NumPy array

# 1. Transpose the image using NumPy
# For a color image (H, W, C) → transpose axes (1, 0, 2) to swap height and width
transposed_image = np.transpose(face_array, (1, 0, 2))

# 2. Display original and transposed images side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(image)                  # PIL Image works directly with imshow
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(transposed_image)
axes[1].set_title("Transposed Image")
axes[1].axis("off")

plt.tight_layout()
plt.show()

# 3. Print the shapes
print("Original image shape: ", face_array.shape)
print("Transposed image shape:", transposed_image.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# === IMPORTANT: Start fresh from the ORIGINAL image ===
# Reload the original image to avoid transposition issues
image = Image.open("/content/sample_data/Face.jpg")
face_array = np.array(image)                # Original shape: (H, W, C)

# 1. Create a copy for masking
masked_image = face_array.copy()

# Get dimensions from the ORIGINAL array
h, w = face_array.shape[:2]
print(f"Image shape: {h} height x {w} width")

# 2. Black out regions
# Top-left 50x50
masked_image[0:50, 0:50] = 0

# Center 100x100
cy, cx = h // 2, w // 2
masked_image[cy-50:cy+50, cx-50:cx+50] = 0

# Bottom-right 50x50
masked_image[h-50:h, w-50:w] = 0

# 3. Display
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(masked_image)
axes[1].set_title("Modified Image (Masked)")
axes[1].axis("off")

plt.suptitle("Exercise 3: Image Masking", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
"""
Exercise 4: Image Cropping and Subsetting
Convert the image in Exercise 1 to a grayscale image.
1. Extract the following regions:
> Top half of the image
> Bottom half of the image
> Centre square region
2. Display each cropped image separately.
3. Print the shape of each cropped image.
4. Compare the dimensions of the cropped images with the original image.
5. Save one of the cropped images as a new image file.
"""
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Exercise 4: Image Cropping and Subsetting

# Convert to grayscale
gray_image = np.array(Image.open("/content/sample_data/Face.jpg").convert("L"))

print("=" * 45)
print("         IMAGE SHAPES COMPARISON")
print("=" * 45)

# Get dimensions
h, w = gray_image.shape

# Original shape
print(f"\nOriginal (grayscale) shape : {gray_image.shape}")

# 1. Extract regions

# Top half
top_half = gray_image[0:h//2, :]
print(f"Top half shape             : {top_half.shape}")

# Bottom half
bottom_half = gray_image[h//2:h, :]
print(f"Bottom half shape          : {bottom_half.shape}")

# Centre square region
cy, cx = h // 2, w // 2
half_sq = min(h, w) // 4     # half side of the square
centre_square = gray_image[cy-half_sq:cy+half_sq, cx-half_sq:cx+half_sq]
print(f"Centre square shape        : {centre_square.shape}")


# 2. Display each cropped image separately
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].imshow(gray_image, cmap="gray")
axes[0, 0].set_title(f"Original Grayscale\n{gray_image.shape}")
axes[0, 0].axis("off")

axes[0, 1].imshow(top_half, cmap="gray")
axes[0, 1].set_title(f"Top Half\n{top_half.shape}")
axes[0, 1].axis("off")

axes[1, 0].imshow(bottom_half, cmap="gray")
axes[1, 0].set_title(f"Bottom Half\n{bottom_half.shape}")
axes[1, 0].axis("off")

axes[1, 1].imshow(centre_square, cmap="gray")
axes[1, 1].set_title(f"Centre Square\n{centre_square.shape}")
axes[1, 1].axis("off")

plt.suptitle("Exercise 4: Image Cropping and Subsetting", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()

# 4. Compare dimensions with original
print("\n" + "=" * 45)
print("         DIMENSION COMPARISON")
print("=" * 45)
print(f"{'Region':<20} {'Shape':<15} {'% of Original'}")
print("-" * 45)

original_pixels = gray_image.size
regions = {
    "Original"      : gray_image,
    "Top Half"      : top_half,
    "Bottom Half"   : bottom_half,
    "Centre Square" : centre_square,
}

for name, region in regions.items():
    pct = (region.size / original_pixels) * 100
    print(f"{name:<20} {str(region.shape):<15} {pct:.1f}%")

# 5. Save one of the cropped images
cropped_to_save = Image.fromarray(top_half)
cropped_to_save.save("top_half_cropped.png")
print("\n Top half saved as 'top_half_cropped.png'")
```

```python
"""
Lab Exercise 5: Image Processing with NumPy
Using the image loaded in Exercise 1:
1. Compute:
> Minimum pixel value
> Maximum pixel value
> Mean pixel value
> Standard deviation of pixel values for each  color channel or layer.
2. Normalize the image to the range [0,1].
3. Create a new image where:
> Pixels greater than the mean are set to 255.
> Pixels less than or equal to the mean are set to 0.
4. Display:
> Original image
> new image
"""

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Make sure we have a NumPy array (from previous exercises)
face_array = np.array(image)   # Convert PIL Image to NumPy array

# 1. Compute Statistics per Color Channel
channels = ["Red", "Green", "Blue"]

print("=" * 55)
print("        PIXEL STATISTICS PER COLOR CHANNEL")
print("=" * 55)
print(f"{'Metric':<25} {'Red':>8} {'Green':>8} {'Blue':>8}")
print("-" * 55)

min_vals  = face_array.min(axis=(0, 1))
max_vals  = face_array.max(axis=(0, 1))
mean_vals = face_array.mean(axis=(0, 1))
std_vals  = face_array.std(axis=(0, 1))

print(f"{'Minimum Pixel Value':<25} {min_vals[0]:>8} {min_vals[1]:>8} {min_vals[2]:>8}")
print(f"{'Maximum Pixel Value':<25} {max_vals[0]:>8} {max_vals[1]:>8} {max_vals[2]:>8}")
print(f"{'Mean Pixel Value':<25} {mean_vals[0]:>8.2f} {mean_vals[1]:>8.2f} {mean_vals[2]:>8.2f}")
print(f"{'Standard Deviation':<25} {std_vals[0]:>8.2f} {std_vals[1]:>8.2f} {std_vals[2]:>8.2f}")
print("=" * 55)

# Overall stats (across all channels)
print(f"\nOverall Min  : {face_array.min()}")
print(f"Overall Max  : {face_array.max()}")
print(f"Overall Mean : {face_array.mean():.2f}")
print(f"Overall Std  : {face_array.std():.2f}")


# 2. Normalize the image to range [0, 1]
normalized_image = face_array / 255.0

print("\n" + "=" * 55)
print("              NORMALIZED IMAGE")
print("=" * 55)
print(f"Min  after normalization : {normalized_image.min():.4f}")
print(f"Max  after normalization : {normalized_image.max():.4f}")
print(f"Mean after normalization : {normalized_image.mean():.4f}")
print(f"Dtype                    : {normalized_image.dtype}")

# 3. Create binary thresholded image using mean
overall_mean = face_array.mean()

# Pixels > mean → 255, pixels <= mean → 0
binary_image = np.where(face_array > overall_mean, 255, 0).astype(np.uint8)

print("\n" + "=" * 55)
print("           BINARY THRESHOLDED IMAGE")
print("=" * 55)
print(f"Threshold (mean)         : {overall_mean:.2f}")
print(f"Pixels set to 255        : {(binary_image == 255).sum()}")
print(f"Pixels set to 0          : {(binary_image == 0).sum()}")
print(f"Binary image dtype       : {binary_image.dtype}")
print(f"Binary image shape       : {binary_image.shape}")

# 4. Display Original and New Image
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].imshow(image)   # Original PIL image for best display
axes[0].set_title(f"Original Image\nShape: {face_array.shape}", fontsize=13)
axes[0].axis("off")

axes[1].imshow(binary_image)
axes[1].set_title(f"Binary Thresholded Image\n(Mean threshold = {overall_mean:.2f})", fontsize=13)
axes[1].axis("off")

plt.suptitle("Exercise 5: Image Processing with NumPy", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```
