```python
#Accessing Elements
"""
Row Indexing
column indexing
slicing
"""
slicing_matrix=np.array([
    [3,2,1],
    [5,3,0],
    [6,7,9]
])
print(slicing_matrix[0]) #First row and all columns
print(slicing_matrix[1]) #Second row and all columns
print(slicing_matrix[2]) #Third row and all columns
print(slicing_matrix[:,-1]) #last column and all rows
print(slicing_matrix[1:,1:]) #last column and all rows
print(slicing_matrix[0:2,0:2]) #First and third rows, including the first two rows
```

```python
#Manipulating elements

slicing_matrix=np.array([
    [3,2,1],
    [5,3,0],
    [6,7,9]
])
slicing_matrix[0]=-1 #this will replace all the first rows in the matrix with -1
print(slicing_matrix)
print("............")
slicing_matrix[1,2]=4 #this will replace the elements in the second row, thrid columns with 4.
print(slicing_matrix)
print("..........")
slicing_matrix[-1,0:2]=[-6,-7] # this will replace the first two columns in the last row with -6 and _7
print(slicing_matrix)
print("..........")
```

```python
#Splitting Arrays
#The splitting function 'split()' divides an array into multiple sub arrays.
arr_split=np.array([[1,2,3,4],
                    [5,6,7,8]])
np.split(arr_split, 2, axis=1)
```

```python
#The vsplit() function splits an array vertically(row-wise)
arr_vsplit=np.array([[1,2],
                     [3,4],
                     [5,6],
                     [7,8]])
np.vsplit(arr_vsplit,2)
```

```python
#The hsplit() function splits an array horizontaly (column-wise)
arr_vsplit=np.array([[1,2,3,4],
                     [5,6,7,8]])
print(np.hsplit(arr_vsplit,2))
```

```python
#Three-Dimensional (3D) Arrays
"""
a 3d array is a NumPy array that stores data in three dimensions of layer.
"""
arr3d=np.array([
  [[1,2,3,2],
  [4,5,6,9],
  [3,5,7,8]],
  [[7,8,9,5],
  [10,11,12,13],
  [13,16,18,12]]
])
print(arr3d.shape)
print(".................")
print(arr3d)

#to access layers of an array
print(arr3d[0])
print("............")
print(arr3d[1])

#to access the first row of the second layer
print(arr3d[1,0])

# to access the second row of the first layer
print(arr3d[0,1])

# to acces the first column from the first layer
print(arr3d[0,:,0])

```

```python
#MATHEMATICAL OPERATION ON 3D ARRAYS
arr3d=np.array([
  [[1,2,3,2],
  [4,5,6,9],
  [3,5,7,8]],
  [[7,8,9,5],
  [10,11,12,13],
  [13,16,18,12]]
])

# Sum across columns (axis=2) within each row of every layer
print(arr3d.sum(axis=2))

# Sum across rows (axis=1) within each layer
print(arr3d.sum(axis=1))

# Sum across layers (axis=0)
print(arr3d.sum)

#sum accross layers and rows simultaneously (axis 0 and 1)
print(arr3d.sum(axis=(0,1)))

#sum across row and columns simultaneously (axis 1 and 2)
print(arr3d.sum(axis=(1,2)))
```
