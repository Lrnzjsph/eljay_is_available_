```python
'''
Rename the columns: Agriculture (% GDP), Education Expenditure (% GDP), Export (% GDP), Education Expenditure (% GDP), Export (% GDP),
Health Expenditure (% GDP), Industry (% GDP), Service (% GDP)  to: agriculture, education_expenditure, export, education_expenditure,
export, health_expenditure, industry, service  respectively.
'''
df_countries = df_countries.rename(columns={"Agriculture_(%GDP)": "Agriculture", "Education_Expenditure_(%GDP)": "Education_Expenditure", "Education_Expenditure_(%GDP)": "Education_Expenditure","Export_(%GDP)": "Export", "Health_Expenditure_(%GDP)": "Health_Expenditure", "Industry_(%GDP)": "Industry", "Service_(%GDP)": "Service"})
print(df_countries)
```

```python
# Convert the year to a date type using an appropriate method.
df_countries['Year'] = pd.to_datetime(df_countries['Year'], format='%Y')
df_countries.head()
```

```python
# Use appropriate method to check missing values and duplicates.
df_countries.duplicated().any()  #to check for duplicate items
```

```python
df_countries.isna().sum() #to check for missing values in the dataset, this also gives the total number of missing variables in the data set and the columns affected.
```

```python
# Delete rows with missing values and duplicates using appropriate method.
df_countries.dropna(inplace=True) #this deletes rows with missing values
```

```python
# Save the dataset to your work directory as “cleaned_data.csv”.

# === SAVE IT AS CSV ===
df_countries.to_csv("cleaned_data.csv", index=False)

print("File saved successfully as 'cleaned_data.csv'!")
```

```python
# Select only the Country Name, Year, and GDP columns.
df_selected = df_countries[['Country Name', 'Year', 'GDP']]
print(df_selected)
```

```python
# Create a new column GDP_in _billions by dividing GDP by 1e9.
df_countries['GDP_in_billions'] = df_countries['GDP'] / 1e9
df_countries.head(10)
```

```python
# Add a column called High_Population to label entries where the population is above 100 million.
df_countries['High_Population'] = df_countries['Population'] > 100000000
df_countries
```

```python
# Create a new column named category that classifies countries based on GDP per capita as follows:
'''Low: GDP per capita < 1,000
Medium: 1,000 ≤ GDP per capita < 10,000
High: GDP per capita ≥ 10,000'''
df_countries['Category'] = pd.cut(df_countries['GDP Per Capita'], bins=[0, 1000, 10000, np.inf], labels=['Low', 'Medium', 'High'])
df_countries.head(10)
```

```python
# Sort the dataset by GDP in descending order to find the highest GDP countries.
df_countries.sort_values(by='GDP Per Capita', ascending=False) #the out put of this code rearranges the indexes of the dataset so we use the reset index function to re-number the data accordingly#
```

```python
#. Arrange countries within the same continent in ascending order of Population Density.
df_countries.sort_values(by='Population Density', ascending=True)
```

```python
# What is the average GDP per capita for each continent?
df_countries.groupby('Continent Name')['GDP Per Capita'].mean()
```

```python
# How many records are there for each country in the dataset?
df_countries['Country Name'].value_counts()
```

```python
## What is the total education expenditure for each continent in the year 2010?
df_countries[df_countries['Year'].dt.year == 2010].groupby('Continent Name')['Education Expenditure'].sum()
```

```python
#Calculate the mean, minimum, and maximum of Health Expenditure (% GDP) grouped by continent name
df_countries.groupby('Continent Name')['Health Expenditure'].agg(['mean', 'min', 'max'])
```

```python
# For each continent, calculate the average GDP Per Capita, but only for countries with Education Expenditure (% GDP) greater than 4%.
df_countries[df_countries['Education Expenditure'] > 4].groupby('Continent Name')['GDP Per Capita'].mean()
```

```python
# Filter countries with a population over 50 million, create a column for GDP in billions, and show only the top 5 highest GDPs.How many unique countries are in the dataset?
df_countries[df_countries['Population'] > 50000000][['Country Name', 'GDP']].sort_values(by='GDP', ascending=False).head(20)
```

```python
# Extract distinct combinations of Country Name and Continent Name.
df_countries[['Country Name', 'Continent Name']].drop_duplicates()
```

```python
# How many distinct years of data are available?
df_countries['Year'].nunique()
```

```python
#. Display the first 10 rows of countries sorted by Population Density for 2020.
# Filter for 2020
df_countries_2020 = df_countries[df_countries['Year'] == 2020]

# Sort by Population Density
sorted_df = df_countries_2020.sort_values(by='Population Density')

# Display first 10 rows
sorted_df.head(10)
```

```python
# Show the last 5 entries of GDP figures for Nigeria.
df_countries[df_countries['Country Name'] == 'Nigeria'].sort_values(by='GDP', ascending=False).tail(5)
```
