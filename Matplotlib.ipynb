```python
'''Absolutely! Think of Matplotlib as a box of crayons for Python. 🎨

Python can do math and store data, but Matplotlib helps you draw pictures of your data so it's easier to understand.
'''
#First, get your crayons. tell python you want to use matplotlib
import matplotlib.pyplot as plt

#matplotlib = the drawing lib
#pyplot = the drawing tool
#plt = a nickname so you don't have to type the long name every time.
```

```python
days = [1,2,3,4]
candies = [2,4,6,8]
plt.plot(days, candies) #this connects the dots
plt.show() #this displays the picture
```

```python
# Add a title
#Every good picture needs a name.

plt.plot(days, candies)
plt.title("my candy collection")
plt.show()
```

```python
#Now let's abes the axis (these are sides of the graph)

plt.plot(days, candies)
plt.title("my candy collection")
plt.xlabel("days")
plt.ylabel("candies")
plt.show()
```

```python
#Change the marker (Markers are the dots)

plt.plot(days, candies, marker='o')
plt.title("my candy collection")
plt.xlabel("days")
plt.ylabel("candies")
plt.show()
```

```python
''' Other types of markers include
  marker='o' #circle
  marker='s' #square
  marker='*' #star
  marker='+' #plus
  marker='x' #x
'''
```

```python
# Bar Charts
friends = ['Tom', 'Jane', 'Mike']
candies = [5,8,3]
plt.bar(friends, candies, color='orange')
plt.title("my candy collection")
plt.xlabel("friends")
plt.ylabel("candies")
plt.show()
```

```python
# Pie chart - Pie charts show parts of a whole
''' Imagine how you spend your day:
  sleeping
  playing
  studying
'''

Activities = ['sleep', 'play', 'study']
hours = [8,10,6]
plt.pie(hours, labels=Activities)
plt.show()
```

```python
#Scatter plots - They show dots without connecting them

x = [1,2,3,4,5]
y = [2,5,3,8,7]
plt.scatter(x,y)
plt.show()
#these are useful when checking whether two things are related
#Eg; Hours studied and test score
```

```python
#Multiple lines - these can be used to connect two things at once.
days = [1,2,3,4]

john = [2,4,6,8]
mary = [1,3,5,7]

plt.plot(days, john)
plt.plot(days, mary)

plt.show()

# Now you can compare john and mary with this chart
```

```python
# Add a legend - a legend tells us which line belongs to who
plt.plot(days, john, label='john')
plt.plot(days, mary, label='mary')
plt.legend()
plt.show()
```

```python
# The magic Recipe- most matplotlib charts follow this pattern:

import matplotlib.pyplot as plt

#using this data
x= ['rice','beans', 'plantain']
y=['mon', 'tue', 'wed']

#draw
plt.plot(x,y)

#decorate it with
plt.title('title')
plt.xlabel('x-label')
plt.ylabel('y-label')

#now display
plt.show()
```

```python
months = ['jan','feb','mar','apr']
money = [10,25,15,40]

plt.bar(months, money)
plt.title('money saved')
plt.xlabel('month')
plt.ylabel('dollars')
plt.show()
```

```python
#1. Multiple plots in the same window (subplots)- think of this as putting several images in one photo frame
fig, axes = plt.subplots(1,2, figsize=(12,4))

#first plot
axes[0].plot([1,2,3], [4,5,6])
axes[0].set_title('plot 1')

#second plot
axes[1].plot([1,2,3], [6,5,4])
axes[1].set_title('plot 2')
plt.show()
```

```python
#Display multiple images side by side
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

img1 = Image.open('/content/sample_data/WhatsApp Image 2026-06-18 at 12.14.46 AM.jpeg')
img2 = Image.open('/content/sample_data/portrait edit.png')

fig, axes = plt.subplots(1,2, figsize=(12,5))
axes[0].imshow(img1)
axes[0].set_title('Sir kings')
axes[0].axis('off')

axes[1].imshow(img2)
axes[1].set_title('My portrait')
axes[1].axis('off')

plt.show()
```

```python
#Create a Grid of Images - for example, 4 images in a 2x2 grid:

fig, axes = plt.subplots(2,2, figsize=(8,8))
for i, ax in enumerate(axes.flat):
  ax.plot([1,2,3], [i+1,i+2,i+3])
  ax.set_title(f"plot{i+1}")
plt.tight_layout()
plt.show()
```
